```python
f = pd.read_excel('../data/raw/airquality/omsz/r-no2.xls', index_col=0)
f.reset_index(inplace=True)
display(f)
```

```python
# Convert xlsx files into pandas dataframe and convert it 

No2_df = pd.DataFrame(columns = ["time","Station","NO2"])
print(len(f.columns))
columns = len(f.columns)
for i in range(1,columns):
    #print(i)
    x = pd.DataFrame()
    columnname = f.columns[i]
    x["NO2"] = f[columnname]
    x["Station"] = columnname
    x["time"] = f["Date & Time"]
    No2_df = No2_df.append(x)

display(No2_df)
```
